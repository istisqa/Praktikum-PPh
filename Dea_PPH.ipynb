```python
import matplotlib.pyplot as plt

# 1. Outlier Detection and Handling
numerical_cols = ['pendapatan_operasional', 'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak', 'pph_badan_normal']
plt.figure(figsize=(12, 6))
df.boxplot(column=numerical_cols)
plt.xticks(rotation=45, ha="right")
plt.title('Box Plots of Numerical Features')
plt.tight_layout()
plt.show()

# Example outlier handling (capping):  Adjust thresholds as needed after observing the box plots
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# 2. Data Type Consistency
for col in ['umur_ekonomis', 'nilai_perolehan', 'penyusutan_garis_lurus', 'nilai_buku_akhir']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 3. Duplicate Rows Removal
print(f"Number of rows before removing duplicates: {len(df)}")
df.drop_duplicates(inplace=True)
print(f"Number of rows after removing duplicates: {len(df)}")

# 4. Data Consistency Checks (Specific Columns)
# Example: check if 'umur_ekonomis' is within a reasonable range (0-50 years)
df = df[(df['umur_ekonomis'] >=0 ) & (df['umur_ekonomis'] <= 50)]
# Add further checks for other columns like 'metode_depresiasi' to make sure all values are valid.

display(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ... (previous code blocks)

# 4. Hypothesis Testing (Example: t-test)
# Example using 'kategori_mesin' (replace with appropriate categorical column)
if 'kategori_mesin' in df.columns:
    group1 = df[df['kategori_mesin'] == True]['pph_badan_normal'].dropna()
    group2 = df[df['kategori_mesin'] == False]['pph_badan_normal'].dropna()

    print(f"Group 1 (kategori_mesin == True): {group1.values}")
    print(f"Group 2 (kategori_mesin == False): {group2.values}")

    if len(group1) > 1 and len(group2) > 1:  # Check for at least two observations in each group
        try:
            t_statistic, p_value = stats.ttest_ind(group1, group2)
            print(f"T-test results for 'pph_badan_normal' between 'kategori_mesin' groups:\nT-statistic: {t_statistic:.3f}\nP-value: {p_value:.3f}")

            alpha = 0.05
            if p_value < alpha:
                print("Reject the null hypothesis. There is a statistically significant difference in 'pph_badan_normal' between the groups.")
            else:
                print("Fail to reject the null hypothesis. There is no statistically significant difference in 'pph_badan_normal' between the groups.")

        except ValueError as e:
            print(f"Error performing t-test: {e}. Trying Mann-Whitney U test.")
            statistic, p_value = stats.mannwhitneyu(group1, group2)
            print(f"Mann-Whitney U Test Results:\nStatistic: {statistic}, P-value: {p_value}")

            alpha = 0.05
            if p_value < alpha:
                print("Reject the null hypothesis. There is a statistically significant difference in 'pph_badan_normal' between the groups.")
            else:
                print("Fail to reject the null hypothesis. There is no statistically significant difference in 'pph_badan_normal' between the groups.")

    else:
        print("One or both groups have fewer than two data points. Mann-Whitney U test cannot be performed.")

else:
    print("Column 'kategori_mesin' not found. Cannot perform hypothesis tests.")

# Finish the task
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ... (previous code blocks)

# 4. Hypothesis Testing (Example: Mann-Whitney U test)
# Example using 'kategori_mesin' (replace with appropriate categorical column)
if 'kategori_mesin' in df.columns:
    group1 = df[df['kategori_mesin'] == True]['pph_badan_normal'].dropna()
    group2 = df[df['kategori_mesin'] == False]['pph_badan_normal'].dropna()

    print(f"Group 1 (kategori_mesin == True): {group1.values}")
    print(f"Group 2 (kategori_mesin == False): {group2.values}")

    if len(group1) > 0 and len(group2) > 0:
        statistic, p_value = stats.mannwhitneyu(group1, group2)
        print(f"Mann-Whitney U Test Results:\nStatistic: {statistic}, P-value: {p_value}")

        alpha = 0.05
        if p_value < alpha:
            print("Reject the null hypothesis. There is a statistically significant difference in 'pph_badan_normal' between the groups.")
        else:
            print("Fail to reject the null hypothesis. There is no statistically significant difference in 'pph_badan_normal' between the groups.")
    else:
        print("One or both groups have no data after removing NaN values. Cannot perform Mann-Whitney U test.")
else:
    print("Column 'kategori_mesin' not found. Cannot perform Mann-Whitney U test.")

# Finish the task
```
