## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Create new features, transform data, and handle remaining missing values as instructed in the subtask.



```python
# Calculate yearly growth rates
for col in ['pendapatan_operasional', 'pendapatan_penjualan', 'laba_bersih_normal']:
    df[f'{col}_growth'] = df.groupby('tahun')[col].pct_change() * 100

# Create financial ratios
df['profit_margin'] = (df['laba_bersih_normal'] / df['pendapatan_penjualan']) * 100

# Lagged feature
df['pendapatan_penjualan_lag1'] = df.groupby('tahun')['pendapatan_penjualan'].shift(1)

# Cumulative sum
df['cumulative_laba_bersih'] = df.groupby('tahun')['laba_bersih_normal'].cumsum()

# Numerical representation for tax holiday status
df['tax_holiday_status'] = df['tax_holiday_status'].map({'Ya': 1, 'Tidak': 0})

# Log transformation for skewed variables
for col in ['pendapatan_operasional', 'pendapatan_penjualan']:
    df[f'{col}_log'] = np.log1p(df[col])  # Use log1p to handle zero values

# Handle any remaining missing values after feature engineering
for col in df.columns:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median()) # Use median for numerical
        else:
            df[col] = df[col].fillna(df[col].mode()[0]) # Use mode for categorical

display(df.head())
display(df.info())
```

## Model training

### Subtask:
Model Training: Time Series Forecasting


**Reasoning**:
Prepare the data, train the model, generate forecasts, and evaluate the model performance.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from prophet import Prophet
import numpy as np

# Prepare the data for Prophet
df_prophet = df[['tahun', 'pendapatan_penjualan']].rename(columns={'tahun': 'ds', 'pendapatan_penjualan': 'y'})
df_prophet['tax_holiday_status'] = df['tax_holiday_status']

# Split the data into training and testing sets
train_data, test_data = train_test_split(df_prophet, test_size=0.2, shuffle=False)

# Train the Prophet model
model = Prophet()
model.add_regressor('tax_holiday_status')  # Add the tax holiday status as a regressor
model.fit(train_data)

# Generate forecasts for the test data
future = model.make_future_dataframe(periods=len(test_data), freq='Y')
future['tax_holiday_status'] = df_prophet['tax_holiday_status']
forecast = model.predict(future)

# Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_true = test_data['y']
y_pred = forecast['yhat'][-len(test_data):]

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")


# Plot the forecast
fig1 = model.plot(forecast)
plt.show()
fig2 = model.plot_components(forecast)
plt.show()
```

**Reasoning**:

```python
# Convert 'tahun' column to datetime objects
df['tahun'] = pd.to_datetime(df['tahun'], format='%Y')

# Prepare the data for Prophet
df_prophet = df[['tahun', 'pendapatan_penjualan']].rename(columns={'tahun': 'ds', 'pendapatan_penjualan': 'y'})
df_prophet['tax_holiday_status'] = df['tax_holiday_status']

# Split the data into training and testing sets
train_data, test_data = train_test_split(df_prophet, test_size=0.2, shuffle=False)

# Train the Prophet model
model = Prophet()
model.add_regressor('tax_holiday_status')  # Add the tax holiday status as a regressor
model.fit(train_data)

# Generate forecasts for the test data
future = model.make_future_dataframe(periods=len(test_data), freq='Y')
future['tax_holiday_status'] = df_prophet['tax_holiday_status']
forecast = model.predict(future)

# Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_true = test_data['y']
y_pred = forecast['yhat'][-len(test_data):]

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")


# Plot the forecast
fig1 = model.plot(forecast)
plt.show()
fig2 = model.plot_components(forecast)
plt.show()
```
