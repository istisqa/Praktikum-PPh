```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

## Data cleaning

### Subtask:
Data Cleaning: Handle potential inconsistencies and errors in the data.


**Reasoning**:
Inspect the data for inconsistencies, errors, or unexpected values, paying close attention to data types. Address inconsistencies, correct data type mismatches, and remove duplicates. Then, verify the cleaning process.



```
# Check for duplicates
duplicates = df.duplicated()
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")
if num_duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicate rows removed.")

# Inspect data types and potential inconsistencies
print(df.dtypes)

# Example cleaning for specific columns (adapt to your actual data)
# Check for non-numeric values in numerical columns
for col in ['tahun', 'pendapatan_operasional', 'pendapatan_penjualan', 'nilai_perolehan', 'umur_ekonomis']:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        if df[col].isnull().any():
            print(f"Non-numeric values found and converted to NaN in column: {col}")
    except KeyError:
        print(f"Column '{col}' not found in DataFrame.")

# Example: Handling outliers (using IQR for 'pendapatan_operasional')
Q1 = df['pendapatan_operasional'].quantile(0.25)
Q3 = df['pendapatan_operasional'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[df['pendapatan_operasional'].between(lower_bound, upper_bound)]
print("Outliers in 'pendapatan_operasional' handled.")

# Verify cleaning
print(df.dtypes)
display(df.describe())
numerical_cols = df.select_dtypes(include=['number']).columns
df[numerical_cols].hist(figsize=(20, 30), bins=10, layout=(10, 4))
plt.tight_layout()
plt.show()
```

## Data analysis

### Subtask:
Analyze the data to identify correlations between variables and explore potential relationships between features and the target variable (if any).  Perform hypothesis testing where appropriate.

Calculate the correlation matrix and visualize it as a heatmap. Then, explore the relationship between 'total_revenue', 'total_expenses', and 'profit' over time using line plots.  Finally, investigate the impact of expense categories on profit using scatter plots and perform hypothesis tests.



```
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Relationship between 'total_revenue', 'total_expenses', and 'profit' over time
plt.figure(figsize=(10, 6))
plt.plot(df['tahun'], df['total_revenue'], label='Total Revenue', color='blue')
plt.plot(df['tahun'], df['total_expenses'], label='Total Expenses', color='red')
plt.plot(df['tahun'], df['profit'], label='Profit', color='green')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Trends of Total Revenue, Total Expenses, and Profit Over Time')
plt.legend()
plt.show()

# Impact of expense categories on profit
expense_categories = ['pembelian_barang', 'beban_operasional', 'biaya_pemeliharaan']
plt.figure(figsize=(15, 5))
for i, category in enumerate(expense_categories):
    plt.subplot(1, 3, i + 1)
    sns.regplot(x=category, y='profit', data=df, color='purple', scatter_kws={'alpha':0.5})
    plt.title(f'Profit vs. {category}')
    plt.xlabel(category)
    plt.ylabel('Profit')
plt.tight_layout()
plt.show()

# Hypothesis Testing (Example: Correlation between 'pembelian_barang' and 'profit')
correlation_coefficient, p_value = stats.pearsonr(df['pembelian_barang'], df['profit'])
print(f"Correlation coefficient between pembelian_barang and profit: {correlation_coefficient}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("There is a statistically significant correlation between pembelian_barang and profit.")
else:
    print("There is no statistically significant correlation between pembelian_barang and profit.")

#More Hypothesis test can be added here if needed.
```
